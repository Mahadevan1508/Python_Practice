```python
#Password Strength Checker with Advanced String Operations
# def check_length(password):
#     if len(password) <= 8:
#         print("Password must be eight characters")
#         return
# print(check_length("Shanparmat"))
# print(check_length("asfhj"))
def check_length(password):
    return len(password) >= 8

print("\nCheck Length")
print(check_length("hello"))        # False
print(check_length("password"))     # True
print(check_length("pass1234"))     # True

def check_case(password):
    has_upper = False
    has_lower = False

    for character in password:
        if character.isupper():
            has_upper = True
        elif character.islower():
            has_lower = True
    return has_upper and has_lower

print("\nCheck Case")
print(check_case("hello"))        # False
print(check_case("PASSWORD"))     # False
print(check_case("PASSword"))     # True

def check_digits(password):
    return any(character.isdigit() for character in password)
print("\nCheck Digits")
print(check_digits("pass123"))
print(check_digits("password"))
print(check_digits("pass@orld"))

def check_special_Chars(password):
    special_chars = {'!','@','#','$','%','^','&','*'} #set
    password_chars = set(password)

    return bool(special_chars.intersection(password_chars))
print('\nCheck Special_Character')
print(check_special_Chars('pass@orld'))
print(check_special_Chars('PASS23'))
print(check_special_Chars('p@$$#or!@'))

def password_strength(password):
    checks = {
        "length" : check_length(password),
        "case"  : check_case(password),
        "digits" : check_digits(password),
        "special": check_special_Chars(password)
    }
    passed = sum(checks.values())        

    total = len(checks)

    if passed == total:
        return "strong"
    elif passed >= total -1:
        return "medium"
    else:
        return "weak"
print('\nPassword Strength')
print(password_strength("Shanparmat@1"))
print(password_strength("shanparmat@1"))
print(password_strength("shan"))
```

```python
#Building a Bank Account System with Classes and Objects
#using instance(object) method
class BankAccount:
    def __init__(self,account_number:int,account_holder:str,balance = 0.0):
        self.account_number = account_number
        self.account_holder = account_holder
        self.__balance = balance #private attribute
        #Deposit Method
    def deposit(self,amount:float):
        if amount <= 0:
            print("Deposit amount must be positive.")
            return

        self.__balance += amount
        print(f"Deposit Successful! New balance is {self.__balance:.2f}")
        #Withdrawal Method
    def withdraw(self,withdraw_amount:float):
        if withdraw_amount <= 0:
            print("withdrawal amount must be positive.")
            return
        if withdraw_amount > self.__balance:
            print("Insufficient Balance")
            return
            
        self.__balance -= withdraw_amount
        print(f"Withdraw Successful! New balance is {self.__balance:.2f}")
    def display_balance(self):
        print(
            f"Account Number: {self.account_number}\n"
            f"Account Holder: {self.account_holder}\n"
            f"Current Balance: {self.__balance:.2f}"
        )
account1 = BankAccount(12345,"Madhan",1000)
account2 = BankAccount(67890,"Santhosh",2000)
account3 = BankAccount(10293,"Prajhwal",3000)
account1.deposit(1000)
account1.withdraw(500)
account1.display_balance()
account2.deposit(-1000)
account2.withdraw(500)
account2.display_balance()
account3.deposit(1000)
account3.withdraw(500)
account3.display_balance()
```

```python
#Inventory System with Magic Methods
class Product:
    def __init__(self,name,price,quantity):
        self.name = name
        self.price = price
        self.quantity = quantity
    def __str__(self):
        return f"Product:{self.name},\nPrice:{self.price},\nQty:{self.quantity}"
    def __add__(self,other):
        if self.name != other.name:
            raise ValueError("Cannot add products with different names.")
        self.quantity += other.quantity
        return Product(self.name, self.price, self.quantity)
    def __lt__(self,other):
        return self.price < other.price
    def __gt__(self,other):
        return self.price > other.price
    def __eq__(self,other):
        return self.price == other.price
      
p1 = Product("Iphone",200000,5)
p2 = Product("Iphone",100000,15)
p3 = Product("Mac",200000,5)
print("Adding Quantities:") 

combined = p1 + p2
print(combined)

print("Less Than:")
print(p1 < p2)

print("Greater Than:")
print(p1 > p2)

print("Equal:")
print(p1 == p2)
print(p1 == p3)
```

```python
#Shape Hierarchy with Abstract Classes
from abc import ABC, abstractmethod
import math
class Shape(ABC):
    @abstractmethod
    def calculate_area(self):
        pass
    @abstractmethod        
    def calculate_perimeter(self):
        pass

class Circle(Shape):
    def __init__(self, radius):
        if radius < 0:
            raise ValueError("radius must be a numeric values")
        self.radius = float(radius)
    def calculate_area(self):
        return 3.14 * self.radius * self.radius
    def calculate_perimeter(self):
        return 2 * 3.14 * self.radius

class Rectangle(Shape):
    def __init__(self,length,width):
        if length < 0 or width < 0:
            raise ValueError("length and width cannot be negative")
        self.length = float(length)
        self.width = float(width)
    def calculate_area(self):
        return self.length * self.width
    def calculate_perimeter(self):
        return 2 * (self.length + self.width)

class Triangle:
    def __init__(self, a, b, c):
        self.a = a
        self.b = b
        self.c = c
        
        # Validate triangle inequality
        if not self.is_valid():
            raise ValueError("Invalid triangle sides")

    def is_valid(self):
        return (
            self.a + self.b > self.c and
            self.a + self.c > self.b and
            self.b + self.c > self.a
        )

    def perimeter(self):
        return self.a + self.b + self.c

    def area(self):
        s = self.perimeter() / 2   # semi-perimeter
        return math.sqrt(s * (s - self.a) * (s - self.b) * (s - self.c))
        
c = Circle(4)
print("Circle")
print("Radius1: ", c.radius)
print("Area: ", c.calculate_area())
print("Perimeter: ", c.calculate_perimeter())
r = Rectangle(2,4)
print("Rectangle")
print("Length & Width: ",r.length,r.width)
print("Area: ", r.calculate_area())
print("Perimeter: ", r.calculate_perimeter())
t = Triangle(5, 6, 7)
print("Triangle")
print("Perimeter:", t.perimeter())
print("Area:", round(t.area(), 2))
```

```python
#Advanced File Handling with Classes
import json
from json import JSONDecodeError
from pathlib import Path
from typing import Dict, List, Optional, Any

class UserDataManager:
    def __init__(self, filename: str):
        self.filepath = Path(filename)
        try:
            if self.filepath.exists():
                if self.filepath.is_dir():
                    raise ValueError(f"Path points to a directory, not a file: {self.filepath}")
                # Validate existing file contains valid JSON list; if empty/invalid, reset to []
                try:
                    with self.filepath.open("r", encoding="utf-8") as f:
                        data = json.load(f)
                    if not isinstance(data, list):
                        with self.filepath.open("w", encoding="utf-8") as f:
                            json.dump([], f, indent=4)
                except JSONDecodeError:
                    with self.filepath.open("w", encoding="utf-8") as f:
                        json.dump([], f, indent=4)
            else:
                # Ensure parent directories exist and create file initialized as []
                self.filepath.parent.mkdir(parents=True, exist_ok=True)
                with self.filepath.open("w", encoding="utf-8") as f:
                    json.dump([], f, indent=4)
        except PermissionError as e:
            raise PermissionError(f"Insufficient permission for path: {self.filepath}") from e
        except ValueError:
            raise
        except OSError as e:
            raise OSError(f"Failed to prepare file at: {self.filepath}") from e

    def read_users(self) -> List[Dict]:
        """
        Reads users from the JSON file. Always returns a list.
        Resets the file to [] if JSON is invalid.
        """
        try:
            with self.filepath.open("r", encoding="utf-8") as f:
                data = json.load(f)
            if isinstance(data, list):
                return data
            print(f"Warning: Expected a list in '{self.filepath}', found {type(data).__name__}. Resetting to [].")
            with self.filepath.open("w", encoding="utf-8") as f:
                json.dump([], f, indent=4)
            return []
        except FileNotFoundError:
            print(f"Error: The file '{self.filepath}' was not found. Recreating as empty list.")
            with self.filepath.open("w", encoding="utf-8") as f:
                json.dump([], f, indent=4)
            return []
        except PermissionError:
            print(f"Error: Permission denied when trying to read '{self.filepath}'.")
            return []
        except JSONDecodeError:
            print(f"Error: The file '{self.filepath}' contains invalid JSON. Resetting to [].")
            with self.filepath.open("w", encoding="utf-8") as f:
                json.dump([], f, indent=4)
            return []
        except Exception as e:
            print(f"An unexpected error occurred while reading '{self.filepath}': {e}")
            return []

    def write_user(self, user_data: Dict) -> None:
        """
        Appends a single user dictionary to the JSON file.
        Reads existing data, updates it, then writes back.
        """
        users = self.read_users()
        users.append(user_data)

        try:
            with self.filepath.open("w", encoding="utf-8") as f:
                json.dump(users, f, indent=4, ensure_ascii=False)
            print(f"User successfully added to '{self.filepath}'.")
        except PermissionError:
            print(f"Error: Cannot write to '{self.filepath}'. Permission denied.")
        except Exception as e:
            print(f"Unexpected error while writing to '{self.filepath}': {e}")

    def update_user(self, email: str, updates: Dict[str, Any], *, case_insensitive: bool = True) -> bool:
        """
        Update an existing user's data identified by their email.
        Returns True if a user was found and updated; False otherwise.
        """
        if not isinstance(email, str) or not email.strip():
            print("Error: A non-empty email string is required.")
            return False

        if not isinstance(updates, dict) or not updates:
            print("Error: 'updates' must be a non-empty dictionary.")
            return False

        users = self.read_users()
        if not users:
            print(f"Validation: No users found in '{self.filepath}'.")
            return False

        target = email.strip()
        found_index: Optional[int] = None

        for idx, user in enumerate(users):
            if not isinstance(user, dict):
                continue
            user_email = user.get("email")
            if not isinstance(user_email, str):
                continue

            if case_insensitive:
                if user_email.strip().lower() == target.lower():
                    found_index = idx
                    break
            else:
                if user_email.strip() == target:
                    found_index = idx
                    break

        # <-- This check must be after the loop
        if found_index is None:
            print(f"Validation: No user found with email '{email}'.")
            return False

        # Do not allow changing the primary key (email)
        updates = {k: v for k, v in updates.items() if k != "email"}
        if not updates:
            print("Info: No updatable fields provided (email cannot be changed). Nothing to update.")
            return False

        # Merge updates (shallow update)
        users[found_index].update(updates)

        # Persist back to file
        try:
            with self.filepath.open("w", encoding="utf-8") as f:
                json.dump(users, f, indent=4, ensure_ascii=False)
            print(f"User with email '{email}' updated successfully.")
            return True
        except PermissionError:
            print(f"Error: Cannot write to '{self.filepath}'. Permission denied.")
        except Exception as e:
            print(f"Unexpected error while writing updates to '{self.filepath}': {e}")
        return False

mgr = UserDataManager("users.json")

# Seed users (with 'email')
mgr.write_user({"id": 3, "name": "Priya", "email": "priya@example.com", "role": "Viewer"})
mgr.write_user({"id": 4, "name": "Arun",  "email": "arun@example.com",  "role": "Editor"})

# Update by email
mgr.update_user("priya@example.com", {"role": "Admin"})                       # True
mgr.update_user("ARUN@example.com", {"name": "Arun Kumar"}, case_insensitive=True)  # True

# Non-existent user
mgr.update_user("ghost@example.com", {"role": "Owner"})                        # False
```

```python
#Context Manager for Database-like File Operations
import csv
from pathlib import Path
from typing import List, Optional, Callable, Dict, Any, Union


class CSVTransaction:
    """
    Context manager for working with a CSV in a transactional, in-memory way.
    - Reads entire CSV into memory (self.data) on enter.
    - Modifications (insert/update/delete) affect only memory by default.
    - Use commit() to persist self.data back to disk.
    """

    def __init__(self, filename: str, headers: Optional[List[str]] = None):
        self.filename = Path(filename)
        self.headers = headers

        self.file = None  # file handle (for reading existing content)
        # In-memory storage
        self.data: List[Dict[str, str]] = []
        self.original_data: List[Dict[str, str]] = []

    def __enter__(self):
        file_exists = self.filename.exists()

        # Open file for read (we won't append here; we keep data in memory)
        mode = "r" if file_exists else "w+"
        self.file = self.filename.open(mode, newline="", encoding="utf-8")

        # If file exists, try to read header; otherwise write header (must be provided)
        if file_exists:
            self.file.seek(0)
            reader = csv.reader(self.file)
            try:
                existing_header = next(reader)
            except StopIteration:
                existing_header = None

            if existing_header:
                self.headers = existing_header
            else:
                # Empty existing file: need headers to initialize
                if not self.headers:
                    raise ValueError("Headers must be provided for an empty CSV file.")
                # Rewrite header
                self.file.close()
                with self.filename.open("w", newline="", encoding="utf-8") as wf:
                    csv.writer(wf).writerow(self.headers)
                # Reopen for reading
                self.file = self.filename.open("r", newline="", encoding="utf-8")
        else:
            # New file: need headers
            if not self.headers:
                raise ValueError("Headers must be provided when creating a new CSV file.")
            # Write header for the new file
            self.file.write(",".join(self.headers) + "\n")
            self.file.flush()
            self.file.seek(0)

        # Load complete CSV into memory
        self.file.seek(0)
        dict_reader = csv.DictReader(self.file)
        self.data = [dict(row) for row in dict_reader]  # values are already strings
        self.original_data = [row.copy() for row in self.data]  # backup snapshot

        return self

    # ---------- Helpers ----------
    def _validate_exact_row(self, row: Dict[str, Any], op: str):
        row_keys = set(row.keys())
        header_keys = set(self.headers or [])
        extra = row_keys - header_keys
        missing = header_keys - row_keys
        if extra:
            raise ValueError(f"[{op}] Unexpected columns: {sorted(extra)}. Expected: {self.headers}")
        if missing:
            raise ValueError(f"[{op}] Missing required columns: {sorted(missing)}. Required: {self.headers}")

    def _validate_subset_columns(self, cols: List[str], op: str):
        header_set = set(self.headers or [])
        unknown = [c for c in cols if c not in header_set]
        if unknown:
            raise ValueError(f"[{op}] Unknown column(s): {unknown}. Expected subset of {self.headers}")

    # ---------- In-memory write ops ----------
    def writerow(self, row: Dict[str, Any]):
        """Alias to insert_row to keep modifications in memory only."""
        self.insert_row(row)

    def writerows(self, rows):
        for r in rows:
            self.insert_row(r)

    def insert_row(self, row: Dict[str, Any]):
        """Insert a new row into memory after schema validation."""
        if not self.headers:
            raise RuntimeError("Headers are not set; cannot validate insert.")
        self._validate_exact_row(row, op="insert")
        # Normalize values to strings
        normalized = {k: "" if row.get(k) is None else str(row[k]) for k in self.headers}
        self.data.append(normalized)

    def update_row(self, where: Union[Callable[[Dict[str, str]], bool], Dict[str, Any]], updates: Dict[str, Any]) -> int:
        """
        Update in-memory rows that match `where` with values from `updates`.
        - where: callable(row)->bool OR dict of exact matches {column: value}
        - updates: dict {column: value}; validated against headers
        Returns: number of rows updated.
        """
        if not self.headers:
            raise RuntimeError("Headers are not set; cannot validate updates.")

        # Validate update columns
        self._validate_subset_columns(list(updates.keys()), op="update")

        # Build predicate from `where`
        if callable(where):
            pred: Callable[[Dict[str, str]], bool] = where
        elif isinstance(where, dict):
            self._validate_subset_columns(list(where.keys()), op="filter")
            needle = {k: "" if v is None else str(v) for k, v in where.items()}
            pred = lambda row: all(row.get(k, "") == needle[k] for k in needle)
        else:
            raise TypeError("`where` must be a callable(row)->bool or a dict of {column: value}")

        # Normalize updates to strings
        updates_norm = {k: "" if v is None else str(v) for k, v in updates.items()}

        count = 0
        for row in self.data:
            if pred(row):
                for k, v in updates_norm.items():
                    row[k] = v
                count += 1
        return count

    def delete_row(
        self,
        where: Union[Callable[[Dict[str, str]], bool], Dict[str, Any]],
        limit: Optional[int] = None,
    ) -> int:
        """
        Delete rows in memory only (does not persist to disk).
        - where: callable(row)->bool OR dict {column: value}
        - limit: optional max number of rows to delete
        Returns: number of rows deleted.
        """
        if not self.headers:
            raise RuntimeError("Headers are not set; cannot validate delete.")

        # Build predicate from `where`
        if callable(where):
            pred: Callable[[Dict[str, str]], bool] = where
        elif isinstance(where, dict):
            self._validate_subset_columns(list(where.keys()), op="filter")
            needle = {k: "" if v is None else str(v) for k, v in where.items()}
            pred = lambda row: all(row.get(k, "") == needle[k] for k in needle)
        else:
            raise TypeError("`where` must be a callable(row)->bool or a dict of {column: value}")

        kept: List[Dict[str, str]] = []
        deleted = 0
        for row in self.data:
            if pred(row) and (limit is None or deleted < limit):
                deleted += 1
                continue
            kept.append(row)
        self.data = kept
        return deleted

    # ---------- Optional persistence ----------
    def commit(self) -> None:
        """Persist current in-memory rows to disk by rewriting the CSV."""
        if not self.headers:
            raise RuntimeError("Headers are not set; cannot commit.")
        with self.filename.open("w", newline="", encoding="utf-8") as f:
            writer = csv.DictWriter(f, fieldnames=self.headers)
            writer.writeheader()
            writer.writerows(self.data)

    def rollback(self) -> None:
        """Restore the original snapshot (discard in-memory changes)."""
        self.data = [row.copy() for row in self.original_data]

    def __exit__(self, exc_type, exc, tb):
        if self.file:
            try:
                self.file.close()
            finally:
                self.file = None
        # Do not auto-commit; caller decides.
        return False
# Create new CSV
with CSVTransaction("employees.csv", headers=["id", "name", "email"]) as txn:
    txn.insert_row({"id": 1, "name": "Madhan", "email": "Madhan@example.com"})
    # Nothing is written yet. To persist this row:
    txn.commit()

# Append to existing CSV (in memory first)
with CSVTransaction("employees.csv") as txn:
    # txn.insert_row({"id": 2, "name": "Mahadevan", "email": "devanmaha@example.com"})
    # txn.insert_row({"id": 3, "name": "Shantanab", "email": "shantanab@example.com"})
    txn.insert_row({"id": 4, "name": "Ananya", "email": "ananya@example.com"})

    # # Update name where id == "2" (values are strings in CSV)
    # txn.update_row(where={"id": "2"}, updates={"name": "Maddy"})

    # Delete by dict predicate (e.g., delete id == "3")
    txn.delete_row(where={"id": "3"})

    # Or delete by callable (e.g., delete name starting with 'A', only 1 row)
    txn.delete_row(where=lambda r: r["name"].startswith("A"), limit=1)

    print("In-memory data:", txn.data)             # shows current staged state
    print("Original snapshot:", txn.original_data) # shows the original loaded state

    # To persist changes:
    txn.commit()
```

```python
#Logging Decorator for File Operations
import json
import logging
import time
from functools import wraps
from logging.handlers import TimedRotatingFileHandler
from pathlib import Path
from typing import Any, Dict, List, Optional

# -------------------------
# Setup Daily Rotating Log
# -------------------------
LOGGER_NAME = "user_data_manager"
LOG_PATH = Path("logs") / "app.log"

def setup_logging():
    logger = logging.getLogger(LOGGER_NAME)
    logger.setLevel(logging.INFO)
    logger.propagate = False

    # Remove existing handlers (important if re-running)
    for h in list(logger.handlers):
        logger.removeHandler(h)
        h.close()

    LOG_PATH.parent.mkdir(parents=True, exist_ok=True)
    handler = TimedRotatingFileHandler(
        filename=str(LOG_PATH),
        when="midnight",     # rotate daily
        interval=1,
        backupCount=7,       # keep last 7 files
        encoding="utf-8",
        delay=True
    )
    formatter = logging.Formatter("%(asctime)s | %(levelname)s | %(name)s | %(message)s",
                                  datefmt="%Y-%m-%d %H:%M:%S")
    handler.setFormatter(formatter)
    logger.addHandler(handler)
    return logger

logger = setup_logging()

# -------------------------
# Decorator (print + file)
# -------------------------
def log_file_operation(operation: str, level=logging.INFO):
    def decorator(func):
        @wraps(func)
        def wrapper(*args, **kwargs):
            start = time.perf_counter()
            msg = f"{operation} started | args={args[1:]}, kwargs={kwargs}"  # skip self in args view
            print(msg)
            logging.getLogger(LOGGER_NAME).log(level, msg)
            try:
                result = func(*args, **kwargs)
                elapsed = time.perf_counter() - start
                msg = f"{operation} successful | time={elapsed:.6f}s"
                print(msg)
                logging.getLogger(LOGGER_NAME).log(level, msg)
                return result
            except Exception as e:
                elapsed = time.perf_counter() - start
                msg = f"{operation} failed | time={elapsed:.6f}s | error={e}"
                print(msg)
                logging.getLogger(LOGGER_NAME).exception(msg)  # includes traceback in file
                raise
        return wrapper
    return decorator

# -------------------------
# Assignment 1: UserDataManager
# -------------------------
class UserDataManager:
    """
    Example implementation:
    - Stores user records as a list of dicts in a JSON file: [{"id": 1, "name": "..."}, ...]
    """
    def __init__(self, json_path: str | Path):
        self.path = Path(json_path)

    def _load_all(self) -> List[Dict[str, Any]]:
        if not self.path.exists():
            return []
        with open(self.path, "r", encoding="utf-8") as f:
            return json.load(f)

    def _save_all(self, items: List[Dict[str, Any]]) -> None:
        self.path.parent.mkdir(parents=True, exist_ok=True)
        with open(self.path, "w", encoding="utf-8") as f:
            json.dump(items, f, ensure_ascii=False, indent=2)

    # Decorated methods:
    @log_file_operation("read_users")
    def read(self) -> List[Dict[str, Any]]:
        """Return all users."""
        return self._load_all()

    @log_file_operation("write_user")
    def write(self, user: Dict[str, Any]) -> Dict[str, Any]:
        """
        Append a new user. If 'id' is missing, assign next int id.
        """
        items = self._load_all()
        if "id" not in user:
            next_id = (max((u.get("id", 0) for u in items), default=0) + 1)
            user = {**user, "id": next_id}
        else:
            # Ensure uniqueness
            if any(u.get("id") == user["id"] for u in items):
                raise ValueError(f"User with id={user['id']} already exists.")
        items.append(user)
        self._save_all(items)
        return user

    @log_file_operation("update_user")
    def update(self, user_id: int, updates: Dict[str, Any]) -> Dict[str, Any]:
        """
        Update fields for a specific user_id.
        """
        items = self._load_all()
        for u in items:
            if u.get("id") == user_id:
                u.update(updates)
                self._save_all(items)
                return u
        raise ValueError(f"User with id={user_id} not found.")

    # (Optional) more decorated operations
    @log_file_operation("delete_user")
    def delete(self, user_id: int) -> None:
        items = self._load_all()
        new_items = [u for u in items if u.get("id") != user_id]
        if len(new_items) == len(items):
            raise ValueError(f"User with id={user_id} not found.")
        self._save_all(new_items)

# -------------------------
# Example usage
# -------------------------
if __name__ == "__main__":
    manager = UserDataManager("data/users.json")

    # Write a user
    manager.write({"name": "Alice", "email": "alice@example.com"})

    # Read all users
    users = manager.read()
    print("Users:", users)

    # Update user
    if users:
        uid = users[0]["id"]
        updated = manager.update(uid, {"email": "alice+upd@example.com"})
        print("Updated:", updated)
```

```python
#File Backup System with Exception Handling
import shutil
from pathlib import Path
from datetime import datetime

class FileBackupManager:
    def __init__(self, target_file, max_backups):
        self.target = Path(target_file)
        self.backup_dir = self.target.parent / "backups"
        self.backup_dir.mkdir(exist_ok=True)
        self.max_backups = max_backups

    # Backup Rotation
    def _rotate_backup(self):
        backups = sorted(self.backup_dir.glob(f"{self.target.name}.*.bak"))
        extra = len(backups) - self.max_backups
        if extra > 0:
            for old in backups[:extra]:
                old.unlink()  # delete oldest backup

    # Create a timestamped backup
    def create_backup(self):
        timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
        backup_name = f"{self.target.name}.{timestamp}.bak"
        backup_path = self.backup_dir / backup_name
        shutil.copy2(self.target, backup_path)

        # ✅ Call rotation here
        self._rotate_backup()

        return backup_path

    # List all backups
    def list_backups(self):
        return sorted(self.backup_dir.glob(f"{self.target.name}.*.bak"))

    # Context manager for safe modification
    def modify_with_backup(self):
        manager = self

        class _Context:
            def __enter__(self):
                manager._backup = manager.create_backup()
                return manager.target
            def __exit__(self, exc_type, exc, tb):
                if exc_type:
                    if manager._backup.exists():
                        shutil.copy2(manager._backup, manager.target)
                return False

        return _Context()

    # Recover by index or timestamp substring
    def recover(self, identifier):
        backups = self.list_backups()
        if not backups:
            print("No backups available.")
            return

        if isinstance(identifier, int):
            if identifier < 0 or identifier >= len(backups):
                print("Invalid backup index.")
                return
            backup_file = backups[identifier]
        else:
            matches = [b for b in backups if identifier in b.name]
            if not matches:
                print("No matching backup found.")
                return
            backup_file = matches[0]

        if not backup_file.exists():
            print("Backup file is invalid.")
            return

        shutil.copy2(backup_file, self.target)
        print(f"Restored from backup: {backup_file}")


if __name__ == "__main__":
    mgr = FileBackupManager("backup.txt", 3)

    # Create demo file if not exists
    Path("backup.txt").write_text("Test")

    # Create multiple backups
    for _ in range(6):
        mgr.create_backup()

    # Print backups after rotation
    print("Backups after rotation:")
    for b in mgr.list_backups():
        print(" -", b.name)
```

```python
# Implementing a Retry Mechanism with Decorators
import time
import requests
import logging
import random  # ✅ FIX 1: Missing import

# Configure logging (simple format with timestamp)
logging.basicConfig(
    level=logging.INFO,
    format="%(asctime)s - %(levelname)s - %(message)s"
)

def retry(max_retries=3, delay=1, exceptions=(requests.exceptions.HTTPError,)):
    def decorator(func):
        def wrapper(*args, **kwargs):
            for attempt in range(1, max_retries + 1):
                try:
                    return func(*args, **kwargs)

                except exceptions as e:
                    # Log the failure
                    logging.warning(f"Attempt {attempt} failed: {e}")

                    if attempt == max_retries:
                        logging.error("Max retries reached. Raising exception.")
                        raise

                    # Exponential backoff
                    wait_time = (2 ** attempt) * delay
                    logging.info(f"Retrying in {wait_time} seconds...")
                    time.sleep(wait_time)
                    
        return wrapper
    return decorator

@retry(
    max_retries=3,
    delay=1,
    exceptions=(requests.exceptions.HTTPError, requests.exceptions.SSLError)
)
def fetch_data(url):
    response = requests.get(url)
    response.raise_for_status()
    return response.text
    
# ✅ Mock API that randomly fails (simulating 500 errors)
@retry(max_retries=3, delay=1)
def mock_api_call():
    # Randomly succeed or fail
    if random.choice([True, False]):
        raise Exception("500 Internal Server Error (simulated)")
    return "Success!"

# ✅ Run the test
if __name__ == "__main__":
    print("Calling Mock API...")
    result = mock_api_call()
    print("Final Result:", result)

    print("Calling fetch_data API...")
    data = fetch_data("https://example.com/bad-url")
    print(data)
```

```python
# Concurrent File Downloader with ThreadPoolExecutor
import os
import logging
import requests
from concurrent.futures import ThreadPoolExecutor, as_completed
from urllib.parse import urlparse
from tqdm import tqdm

DOWNLOAD_DIR = "./downloads"
LOG_FILE = "errors.log"
CHUNK_SIZE = 8192  # 8 KB

# Configure logging
logging.basicConfig(
    filename=LOG_FILE,
    level=logging.ERROR,
    format="%(asctime)s - %(levelname)s - %(message)s",
)


def download_file(url):
    """
    Download a single file.
    Returns (filename, bytes_downloaded) on success.
    Raises exception on failure.
    """
    os.makedirs(DOWNLOAD_DIR, exist_ok=True)

    filename = os.path.basename(urlparse(url).path) or "downloaded_file"
    file_path = os.path.join(DOWNLOAD_DIR, filename)

    response = requests.get(url, timeout=5, verify=True, stream=True)
    response.raise_for_status()

    bytes_downloaded = 0
    with open(file_path, "wb") as f:
        for chunk in response.iter_content(chunk_size=CHUNK_SIZE):
            if chunk:
                f.write(chunk)
                bytes_downloaded += len(chunk)

    return filename, bytes_downloaded


def download_all(urls):
    """Download multiple files concurrently with progress tracking and error logging."""
    with ThreadPoolExecutor(max_workers=5) as executor:
        futures = {executor.submit(download_file, url): url for url in urls}

        with tqdm(total=len(urls), desc="Files completed", unit="file") as progress:
            for future in as_completed(futures):
                url = futures[future]
                try:
                    filename, size = future.result()
                    progress.write(f"Downloaded {filename} ({size} bytes)")
                except requests.exceptions.RequestException as e:
                    # Log known network-related errors
                    logging.error("Failed to download %s: %s", url, e)
                    progress.write(f"Skipped {url} (network error)")
                except Exception as e:
                    # Catch-all for unexpected errors
                    logging.error("Unexpected error for %s: %s", url, e)
                    progress.write(f"Skipped {url} (unexpected error)")
                finally:
                    progress.update(1)


if __name__ == "__main__":
    urls = [
        "https://www.rfc-editor.org/rfc/rfc2616.txt",
        "https://www.rfc-editor.org/rfc/rfc7231.txt",
        "https://invalid.example.com/file.txt",  # will fail & be logged
        "https://www.rfc-editor.org/rfc/rfc7540.txt",
    ]

    download_all(urls)
```

```python
#Regex-Based Data Masking for PII
import re
import timeit

# ----------------------------
# PII Detection Patterns
# ----------------------------
PII_PATTERNS = {
    "email": re.compile(
        r"[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Za-z]{2,}"
    ),
    "india_phone": re.compile(
        r"(?<!\d)\+91-\d{10}(?!\d)"
    ),
    "credit_card": re.compile(
        r"(?:\d[ -]*?){13,16}"
    ),
}

sample_text = """
Contact us at support@example.com or admin.test@company.co.
Call +91-9876543210 for assistance.
Payment cards: 4111 1111 1111 1111, 5500-0000-0000-0004.
"""

# ----------------------------
# Show detected PII
# ----------------------------
for pii_type, pattern in PII_PATTERNS.items():
    print(f"{pii_type.upper()} matches:")
    for match in pattern.findall(sample_text):
        print(f"  - {match}")
    print()


def mask_pii(text: str) -> str:
    # Mask email (keep first 2 chars)
    text = re.sub(
        r"([A-Za-z0-9._%+-]{2})[A-Za-z0-9._%+-]*(@[A-Za-z0-9.-]+\.[A-Za-z]{2,})",
        r"\1***\2",
        text,
    )

    # Mask Indian phone numbers
    text = re.sub(
        r"(\+91-)\d{6}(\d{4})",
        r"\1******\2",
        text,
    )

    # Mask credit cards (keep last 4 digits)
    text = re.sub(
        r"(?:\d[ -]*?){12}(\d{4})",
        r"**** **** **** \1",
        text,
    )

    return text

def mask_pii_batch(multiline_text: str) -> str:
    lines = multiline_text.splitlines(keepends=True)
    masked_lines = [mask_pii(line) for line in lines]
    return "".join(masked_lines)

print("Masked output:")
print(mask_pii(sample_text))

print("Multi Line Masked output:")
print(mask_pii_batch(sample_text))

target_size = 10 * 1024 * 1024  # 10 MB
data = sample_text * (target_size // len(sample_text))


elapsed = timeit.timeit(lambda: mask_pii_batch(data),number=1)

print(f"Total time for 2 runs: {elapsed:.2f} seconds")
print(f"Average per run: {elapsed / 2:.2f} seconds")
```

```python
#Async Weather API Fetcher with asyncio
import asyncio
import aiohttp

API_KEY = "06655a5ede9daa049ef022ce09138bca"
BASE_URL = "https://api.openweathermap.org/data/2.5/weather"

semaphore = asyncio.Semaphore(5)

async def fetch_weather(session, city):
    async with semaphore:
        params = {
            "q": city,
            "appid": API_KEY,
            "units": "metric"
        }

        async with session.get(BASE_URL, params=params) as response:
            response.raise_for_status()
            data = await response.json()

            return city, {
                "temp": data["main"]["temp"],
                "humidity": data["main"]["humidity"]
            }

async def main():
    cities = [
        "London", "Paris", "Berlin", "Rome",
        "Madrid", "Vienna", "Prague", "Amsterdam"
    ]

    async with aiohttp.ClientSession() as session:
        tasks = [fetch_weather(session, city) for city in cities]

        aggregated = {}

        # ✅ Process results as they complete
        for completed in asyncio.as_completed(tasks):
            city, weather = await completed
            aggregated[city] = weather
            print(f"{city}: {weather['temp']}°C, {weather['humidity']}%")

    return aggregated

# ✅ Jupyter
result = await main()
```
